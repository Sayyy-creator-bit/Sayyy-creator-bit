#PILAS
- Estructura de datos lineal
- Protocolo de acceso de tipo LIFO (Last In, First Out)
- Los elementos entran y salen desde el mismo extremo: Variable llamada ▶ *Top*
- La *base* es opuesta al Top
- Agregar elementos a la pila ▶ *push* 
- Sacar elementos de la pila ▶ *pop* 

###Métodos

Captura de pantalla 2021-11-04 a la(s) 11.59.42.png


```python
class StackADT:
    def __init__(self):
        self.__data = list()

    #True = lista vacía
    def is_empty(self):
        return len(self.__data) == 0
    
    def get_lenght(self):
        return len(self.__data)

    def pop(self):
        if self.is_empty():
            print("Pila vacía, nada que sacar")
        else:
            return self.__data.pop()
    
    def peek(self):
        if self.is_empty():
            print("Pila vacía, nada que sacar")
        else:
            return self.__data[-1]
    
    def push(self, value):
        self.__data.append(value)
    
    def to_string(self):
        for item in self.__data[::-1]:
            print(f" | { item } | ")
```

```python
pila = StackADT()

print("Está vacía?", pila.is_empty())
print("# elementos:", pila.get_lenght())

pila.push(10)
print("Está vacía?", pila.is_empty())
print("# elementos:", pila.get_lenght())

pila.push(20)
pila.push(30)
pila.push(40)
pila.to_string()

print(pila.pop())

print(pila.peek())
pila.to_string()
```
