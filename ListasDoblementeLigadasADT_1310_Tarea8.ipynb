#Listas Doblemente Ligadas

Estructura de datos ligada, que tiene doble nodo de puntero, que apuntan al siguiente y al anterior nodo.

- Apuntador que marca el punto de entrada(head)
- Apuntador que marca el final(tail) que tambièn es otro punto de entrada
- Permite el recorrido completo en ambas direcciones (transversal)

```python
class NodoDoble:
    def __init__(self, value, siguiente=None, previo=None):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

```python
#Ejercicio 1
print("Ejercicio 1")
head = NodoDoble(10)
tail = head
print( head.data )
print( tail.data )

#Ejercicio 2
print("Ejercicio 2")
head = NodoDoble (10, NodoDoble(20))
head.next.prev = head
tail = head.next

print("El primero:", head.data)
print("El primero:", tail.prev.data)


"""
Lo mismo pero con cuatro nodos 10,20,30,40
Imprimir el valor del primero
Imprimir el valor del último
"""

print("Ejercicio 3")
head = NodoDoble(10, NodoDoble(20, NodoDoble(30, NodoDoble(40))))
tail = head.next.next.next
head.next.prev = head
head.next.next.prev = head.next
tail.prev = head.next.next
print("El primero:", head.data)
print("El primero:", tail.prev.prev.prev.data)

print("El último:", tail.data)
print("El último:", head.next.next.next.data)

#Transversal
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

#Reverse transversal
curr_node = tail
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")

print("Agregar 50 al final")
tail.next = NodoDoble( 50, None, tail )
tail = tail.next

#Transversal 2
curr_node = head
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.next
print("")

#Reverse transversal 2
curr_node = tail
while curr_node != None:
  print(f"|{curr_node.data}|-->", end="")
  curr_node = curr_node.prev
print("")
```

```python
class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def get_size(self):
        return self.size
    
    def is_empty(self):
        return self.size == 0

    def append(self, value):
        if self.is_empty():
            self.head = NodoDoble(value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value, None, tail)
            self.tail = self.tail.next
        self.size += 1

    def transversal(self):
        curr_node = self.head
        while curr_node != None:
            print(f"| {curr_node.data} | -->" , end="")
            curr_node = curr_node.next 
        print("")
    
    def find_from_tail(self, value):
        curr_node = self.tail
        conta = 1
        if self.is_empty():
            print("No existen nodos")
        else:
            while curr_node.data != value:
                curr_node = curr_node.prev
                conta += 1
            if curr_node.data == value:
                return conta

    def remove_from_tail(self, value):
        curr_node = self.tail
        while curr_node.data != value:
            curr_node = curr_node.prev
        if curr_node.data == value:   
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1

    def find_from_head(self, value):
        curr_node = self.head
        conta = 1
        if self.is_empty():
            print("No existen nodos")
        else:
            while curr_node.data != value:
                curr_node = curr_node.next
                conta += 1
            if curr_node.data == value:
                return conta
                
    def remove_from_head(self, value):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next
        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1

    def reverse_transversal(self):
            curr_node = self.tail
            while curr_node != None:
                print(f"| {curr_node.data} | -->", end="")
                curr_node = curr_node.prev
            print("")                
```

```python
ldl = DoubleLinkedList()

ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)

print(f"{ldl.get_size()}")
print(f"Está vacía? {ldl.is_empty()}")

print(f"{ldl.find_from_head(30)}")
print(f"{ldl.find_from_tail(20)}")

ldl.remove_from_head(40)
ldl.remove_from_tail(20)

ldl.transversal()
ldl.reverse_transversal()
```
