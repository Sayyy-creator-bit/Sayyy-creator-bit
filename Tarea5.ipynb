```python
class Array:

    def __init__(self, tam):
        #Generador
        self.__datos = [0 for x in range(tam)]  

    def get_item(self, index):
        dato = 0
        if index >= 0 and index < len(self.__datos):
            dato = self.__datos[index]
        else:
            print("Error en el indice")
        return dato

    def set_item(self, dato, index):
        if index >= 0 and index < len(self.__datos):
            self.__datos[index] = dato
        else:
            print("Error en el indice")

    def get_Length(self):
        return len(self.__datos)

    def clear(self, dato):
        self.__datos = [dato for x in range(tam)]  

    def __iter__(self):
        return _getIterator(self.__datos)

class _getIterator:
    def __init__(self, arreglo):
        self.__arreglo = arreglo
        self.__indice_actual = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.__indice_actual < len(self.__arreglo):
            dato = self.__arreglo[self.__indice_actual]
            self.__indice_actual += 1
            return dato
        else:
            raise StopIteration

    def getIterator(self):
        return _getIterator(self.__datos)
```

```python
class Trabajador:
    def __init__(self, nt, nom, apat, amat, he, sb, ai):
        self.__numero_trabajador = nt
        self.__nombre = nom
        self.__paterno = apat
        self.__materno = amat
        self.__horas_extra = he
        self.__sueldo_base = sb
        self.__anio_ingreso = ai 

#métodos de acceso

    def set_nt (self, nt):
        self.__numero_trabajador 
    def get_nt (self):
        return self.__numero_trabajador

    def set_nom (self, nom):
        self.__nombre = nom
    def get_nom (self):
        return self.__nombre 
    def set_apat (self, apat):
        self.__paterno  = apat
    def get_apat (self):
        return self.__paterno

    def set_amat (self, amat):
        self.__materno = amat
    def get_amat (self):
        return self.__materno 

    def set_he (self, he):
        self.__horas_extra = he
    def get_he(self):
        return self.__horas_extra 
    
    def set_sb (self, sb):
        self.__sueldo_base = sb
    def get_sb (self):
        return self.__sueldo_base 
    
    def set_ai (self, ai):
        self.__anio_ingreso = ai
    def get_ai (self):
        return self.__anio_ingreso 

    #métodos de uso general = métodos que hacen cálculos de la lógica del negocio

    def sueldos (self):
        antiguedad = ((2021 - self.__anio_ingreso * 0.3) * self.__sueldo_base)
        horas_ex = 276.5 * self.__horas_extra
        return self.__sueldo_base + antiguedad + horas_ex 
        
    
```

```python
class NominaADT:
    def __init__(self):
        self.__data = []

    #Abrir el archivo y determinar el numero de empleados

    def cargar_datos (self, ruta_archivo):
        archivo = open(ruta_archivo, "rt", encoding='latin -1')
        lineas = archivo.readlines()
        archivo.close()
        
        for iter in range(1,len(lineas)):
              temp = lineas[iter].split(",")
              self.__data.append(Trabajador(temp[0],temp[1],temp[2],temp[3],temp[4],temp[5],temp[6]))
 

    def calcular_sueldos(self):
        #iterar sobre el Array y hacer el calculo de sueldos
        print("TRABAJADORES:\n ")
        
        for emp in self.__data:
          ai=int(emp.get_ai())
          sb=int(emp.get_sb())
          he=int(emp.get_he())
          antiguedad = (((2021 - ai) * 0.3) * sb)
          horas_ex = 276.5 * he
          sueldo= sb+antiguedad+horas_ex
          
          #print(f"El sueldo de {emp.get_nom()} {emp.get_apat()} {emp.get_amat()}  es = {emp.get_sb()}") solo da el sueldo base
          print(f"El sueldo de {emp.get_nom()} {emp.get_apat()} {emp.get_amat()}  es = {sueldo}")
            

    def mas_y_menos_antiguo(self):
        #iterar y buscar el de mayor antiguedad y menor antiguedad
        mayor = 0
        años=[]
        for emp in self.__data:
            ai=int(emp.get_ai())
            antiguedad = 2021 - ai
            años.append(antiguedad)
        print("\nTRABAJADORES CON MAS ANTIGUEDAD:\n")
        mayor=max(años,key=int)
        for emp in self.__data:
            ai=int(emp.get_ai())
            antiguedad = 2021 - ai
            if antiguedad >= mayor:
              print(f"El empleado: {emp.get_nom()} {emp.get_apat()} {emp.get_amat()}, tiene {antiguedad} años de antiguedad")
        minimo=min(años,key=int)
        print("\nTRABAJADORES CON MENOS ANTIGUEDAD:\n")
        for emp in self.__data:
            ai=int(emp.get_ai())
            antiguedad = 2021 - ai
            if antiguedad == minimo:
              print(f"El empleado: {emp.get_nom()} {emp.get_apat()} {emp.get_amat()}, tiene {antiguedad} años de antiguedad")

resultado = NominaADT()
resultado.cargar_datos('junio.dat')
resultado.calcular_sueldos()
print("\n")
resultado.mas_y_menos_antiguo()
#print()
#resultado.menos_antiguo()
```
